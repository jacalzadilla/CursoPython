```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
path=os.getcwd()
data_path = os.path.join(path, 'data')
df=pd.read_csv(os.path.join(data_path, 'Terminos_lagoon_TA_DIC_2023_RawData.csv'))
df.head()
```

```python
sal_data = df["sal_psu"]

plt.hist(sal_data, bins=10)
plt.xlabel("Salinity (PSU)")
plt.ylabel("Probablity")
plt.show()
```

```python
data = df["sal_psu"]


plt.hist(data, bins=10, density=True)
plt.xlabel("Salinity (PSU)")
plt.ylabel("Probablity")
plt.show()
```

```python
do_data = df["do_mg_l"]
plt.boxplot(do_data)

plt.xlabel("DO (mg/L)")
plt.ylabel("Concentration")
plt.title("Boxplot of DO Concentration")

# show plot
plt.show()
```

```python
data = df["sal_psu"]

fig2, ax2 = plt.subplots()

ax2.set_title('Notched boxes')
ax2.boxplot(data, notch=True)

plt.ylabel('Salinity ($PSU$)', fontsize = 12)
plt.show()
```

```python
green_diamond = dict(markerfacecolor='g', marker='D')
fig3, ax3 = plt.subplots()

ax3.set_title('Changed Outlier Symbols')

ax3.boxplot(data, flierprops=green_diamond)


plt.ylabel('TA ($\mu mol  \; kg^{-1}$)', fontsize = 12)
plt.show()
```

### Ejercicios de tarea

```python
fig, ax = plt.subplots(1,2, figsize=(12,6))
ax[0].hist(df['ta_micromol_kg'], bins=15, color='skyblue', edgecolor='black')
ax[0].set_title('Histogram of TA')
ax[0].set_xlabel('TA ($\mu mol  \; kg^{-1}$)')
ax[0].set_ylabel('Frequency')

ax[1].hist(df['dic_micromol_kg'], bins=15, color='skyblue', edgecolor='black')
ax[1].set_title('Histogram of DIC')
ax[1].set_xlabel('DIC ($\mu mol  \; kg^{-1}$)')
ax[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
