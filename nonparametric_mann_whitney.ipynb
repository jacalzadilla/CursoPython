## Mann-Whitney U Test

```python
import pandas as pd
import numpy as np
from scipy.stats import shapiro, mannwhitneyu
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
path=os.getcwd()
data_path = os.path.join(path, 'data')
df=pd.read_csv(os.path.join(data_path, 'Terminos_lagoon_TA_DIC_2023_RawData.csv'))

df.columns
```

```python
def mann_whitney_test(group1, group2, alpha=0.05):
    result = mannwhitneyu(group1, group2, alternative='two-sided')

    print(f'Statistic: {result.statistic}, p-value: {result.pvalue}')
    if result.pvalue < alpha:
        print("Reject the null hypothesis: The two groups have different distributions.")
    else:
        print("Fail to reject the null hypothesis: The two groups have similar distributions.")
```

```python
sns.boxplot(x='season', y='ta_micromol_kg', data=df)
plt.title('Boxplot of TA by Season')
plt.xlabel('Season')
plt.ylabel('TA (micromol/kg)')
plt.show()
```

```python
ta_dry = df[df['season'] == 'Dry']['ta_micromol_kg']
ta_rainy = df[df['season'] == 'Rainy']['ta_micromol_kg']

mann_whitney_test(ta_dry, ta_rainy)
```

```python
sns.boxplot(x='season', y='sal_psu', data=df)
sns.swarmplot(x='season', y='sal_psu', data=df, color=".25")
plt.title('Boxplot of Salinity by Season')
plt.xlabel('Season')
plt.ylabel('Salinity (PSU)')
plt.show()
```

```python
sal_dry = df[df['season'] == 'Dry']['sal_psu']
sal_rainy = df[df['season'] == 'Rainy']['sal_psu']

mann_whitney_test(sal_dry, sal_rainy)
```

```python
sns.violinplot(x='season', y='sal_psu', data=df)
plt.title('Violin Plot of Salinity by Season')
plt.xlabel('Season')
plt.ylabel('Salinity (PSU)')
plt.show()
```
